```python
ft_df[['MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5']]=ft_df[['MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5']].fillna(0)
ft_df['CPI']=ft_df['CPI'].fillna(ft_df['CPI'].mean())
ft_df['Unemployment']=ft_df['Unemployment'].fillna(ft_df['Unemployment'].mean())
```

```python
ft_df.isnull().sum()
```

```python
#convert feature date columns to date time format
ft_df['Date'] = pd.to_datetime(ft_df['Date'],format='%d/%m/%Y')
sales_df['Date']=pd.to_datetime(sales_df['Date'],format='%d/%m/%Y')
```

```python
#Merge store and date column in each sales and features dataframe
#then merge with stores dataframe 
merge_df=sales_df.merge(ft_df,on=['Store','Date','IsHoliday'],how='left')
allmerge_df=merge_df.merge(stores_df, on='Store', how='left')
```

```python
allmerge_df
```

```python
allmerge_df['Date']=pd.to_datetime(merge_df['Date'])
```

```python
#extracting the time components
allmerge_df['Year']=merge_df['Date'].dt.year
allmerge_df['Month']=merge_df['Date'].dt.month
allmerge_df['Week']=merge_df['Date'].dt.isocalendar().week
allmerge_df['DayOfWeek']=merge_df['Date'].dt.dayofweek

```

```python
allmerge_df.drop('Date',axis=1,inplace=True)
```

```python
allmerge_df['Type']=allmerge_df['Type'].replace({"A":1,"B":2,"C":3})
```

```python
allmerge_df['IsHoliday']=allmerge_df['IsHoliday'].apply(lambda x:1 if x==True else 0)
```
