Handling missing Values

```python
ft_df['CPI'].fillna(ft_df['CPI'].median(),inplace=True)
ft_df['Unemployment'].fillna(ft_df['Unemployment'].median(),inplace=True)
```

```python
for i in range(1,6):
  ft_df["MarkDown"+str(i)] = ft_df["MarkDown"+str(i)].apply(lambda x: 0 if x < 0 else x)
  ft_df["MarkDown"+str(i)].fillna(value=0,inplace=True)
```

```python
#convert feature date columns to date time format
ft_df['Date'] = pd.to_datetime(ft_df['Date'],format='%d-%m-%Y')
sales_df['Date']=pd.to_datetime(sales_df['Date'],format='%d/%m/%Y')
```

```python
#Merge store and date column in each sales and features dataframe
#then merge with stores dataframe 
merge_df=sales_df.merge(ft_df,on=['Store','Date','IsHoliday'],how='left')
allmerge_df=merge_df.merge(stores_df, on='Store', how='left')
```

```python
allmerge_df
```

```python
allmerge_df ['Date']=pd.to_datetime(merge_df['Date'])
```

```python
allmerge_df['Year']=merge_df['Date'].dt.year
allmerge_df['Month']=merge_df['Date'].dt.month
allmerge_df['Week']=merge_df['Date'].dt.isocalendar().week
```

```python
allmerge_df
```

```python
allmerge_df.info()
```

```python
allmerge_df['Year'] = allmerge_df['Year'].astype('int64')
allmerge_df['Month'] = allmerge_df['Month'].astype('int64')
allmerge_df['Week'] = allmerge_df['Week'].astype('int64')
```

```python
allmerge_df.drop('Date',axis=1,inplace=True)
```

```python
allmerge_df['Type']=allmerge_df['Type'].replace({"A":1,"B":2,"C":3})
```

```python
allmerge_df['IsHoliday']=allmerge_df['IsHoliday'].apply(lambda x:1 if x==True else 0)
```

```python
#visualize and analyze your dataset
#Time Series Plots for Sales and Markdowns


import matplotlib.pyplot as plt

# Load your data
allmerge_df.to_csv(r"C:\Guvi project\Final Project\ANN Predictive Modeling of Retail Sales and Markdown Impact\cleared_data")# 
# Plot sales over time
plt.figure(figsize=(14, 7))
plt.plot(allmerge_df.index, allmerge_df['Weekly_Sales'], label='Weekly Sales', color='blue')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Weekly Sales Over Time')
plt.legend()
plt.grid(True)
plt.show()

# Plot markdowns over time
markdowns = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']
plt.figure(figsize=(14, 7))
for md in markdowns:
    plt.plot(allmerge_df.index, allmerge_df[md], label=md)
plt.xlabel('Date')
plt.ylabel('Markdown Amount')
plt.title('Markdowns Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot boxplot for sales during holiday vs. non-holiday weeks
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.boxplot(x='IsHoliday', y='Weekly_Sales', data=allmerge_df)
plt.xlabel('Is Holiday')
plt.ylabel('Weekly Sales')
plt.title('Comparison of Sales During Holiday and Non-Holiday Weeks')
plt.show()
```
